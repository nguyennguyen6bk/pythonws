```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

# model = "codellama/CodeLlama-7b-hf"
# model = "gpt2"
model = "codellama/CodeLlama-7b-Python-hf"

# Cấu hình PyTorch để sử dụng GPU nếu có
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

tokenizer = AutoTokenizer.from_pretrained(model)
model = AutoModelForCausalLM.from_pretrained(model)
model.to(device)  # Chuyển mô hình lên GPU (nếu có)

pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,  # Thêm tokenizer vào đây
    device=0 if torch.cuda.is_available() else -1,  # Sử dụng GPU nếu có, nếu không, sử dụng CPU
)

sequences = pipeline(
    'import socket\n\ndef ping_exponential_backoff(host: str):',
    do_sample=True,
    top_k=10,
    temperature=0.1,
    top_p=0.95,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=200,
)
for seq in sequences:
    print(f"Result: {seq['generated_text']}")

# Đảm bảo bạn đã cài đặt thư viện "transformers" phiên bản mới nhất để sử dụng GPU.
# Sử dụng lệnh sau để cài đặt phiên bản mới nhất:
# pip install transformers --upgrade
```
