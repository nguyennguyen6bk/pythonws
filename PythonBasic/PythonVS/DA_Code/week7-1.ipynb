```python
import numpy as np
import pandas as pd

import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# raw_df = pd.read_excel('/Users/phu/Documents/K_31/EDA_K31/data/Career Mode female player datasets - FIFA 16-22.xlsx', sheet_name='FIFA 22')
df = pd.read_csv('/Users/phu/Documents/K_31/EDA_K31/data/players_22.csv')
df
```

```python
cols = ['wage_eur'] # one or more

Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1

new_df = df[~((df[cols] < (Q1 - 1.5 * IQR)) |(df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
new_df = new_df.dropna(subset=['wage_eur'])
```

```python
new_df.wage_eur.isna
```

```python
results = smf.ols("wage_eur ~ potential + overall ", data=new_df).fit()
print(results.summary())
```

```python
results = smf.ols("wage_eur ~ potential + overall + international_reputation", data=new_df).fit()
print(results.summary())
```

```python
results = smf.ols("wage_eur ~ potential + overall + value_eur", data=new_df).fit()
print(results.summary())
```
