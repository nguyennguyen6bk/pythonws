```python
import numpy as np
import pandas as pd

import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df = pd.read_excel('/Users/phu/Documents/K_31/EDA_K31/data/data-vn.xlsx', sheet_name='Data')
df
```

```python
results = smf.ols("Q('Chi tiêu') ~ Q('Thu nhập') + Q('Số người')", data=df).fit()
print(results.summary())
```

```python
results = smf.ols("Q('Chi tiêu') ~ Q('Thu nhập')", data=df).fit()
print(results.summary())
```

```python
results = smf.ols("Q('Chi tiêu') ~ Q('Số người')", data=df).fit()
print(results.summary())
```

```python
df["Pseudo Khu vực"] = df["Khu vực"].map(lambda x: 1 if x == 'Thành thị' else 0)
df
```

```python
results = smf.ols("Q('Chi tiêu') ~ Q('Thu nhập') + Q('Số người') + Q('Pseudo Khu vực')", data=df).fit()
print(results.summary())
```

```python
results = smf.ols("Q('Chi tiêu') ~ Q('Thu nhập') + Q('Pseudo Khu vực')", data=df).fit()
print(results.summary())
```

```python
results = smf.ols("Q('Chi tiêu') ~ Q('Thu nhập') * Q('Pseudo Khu vực')", data=df).fit()
print(results.summary())
```

```python
results = smf.ols("Q('Chi tiêu') ~ Q('Thu nhập') * Q('Pseudo Khu vực') + Q('Số người')", data=df).fit()
print(results.summary())
```
