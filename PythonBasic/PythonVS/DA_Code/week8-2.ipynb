```python
import numpy as np
import pandas as pd

import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df = pd.read_excel('/Users/phu/Documents/K_31/EDA_K31/data/data-vn.xlsx', sheet_name='Data')
df
```

```python
df['khu_vuc'] = df['Khu vực'].map(lambda x: 1 if x == 'Thành thị' else 0)
df
```

```python
results = smf.ols("Q('Chi tiêu') ~ Q('Thu nhập') + Q('Số người') + khu_vuc", data=df).fit()
print(results.summary())
```

```python
results = smf.ols("Q('Chi tiêu') ~ np.log(Q('Thu nhập')) + Q('Số người') + khu_vuc", data=df).fit()
print(results.summary())
```

```python
results = smf.ols("np.log(Q('Chi tiêu')) ~ Q('Thu nhập') + Q('Số người') + khu_vuc", data=df).fit()
print(results.summary())
```

```python
results = smf.ols("np.log(Q('Chi tiêu')) ~ np.log(Q('Thu nhập')) + Q('Số người') + khu_vuc", data=df).fit()
print(results.summary())
```

```python
df['X'] = df['Thu nhập'] + df['Số người']
df
```

```python
results = smf.ols("Q('Chi tiêu') ~ Q('Thu nhập') + Q('Số người') + khu_vuc + X", data=df).fit()
print(results.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# VIF dfframe
new_df = df.pop('Khu vực')

vif_df = pd.DataFrame()
vif_df["feature"] = df.columns

# calculating VIF for each feature
vif_df["VIF"] = [variance_inflation_factor(df.values, i)
						for i in range(len(df.columns))]

print(vif_df)
```
