```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv("data-vn.csv")
df
```

```python
df.rename(columns={"Chi tiêu": "chi_tieu", "Thu nhập": "thu_nhap"}, inplace = True)
df
```

```python
df = df.iloc[:, lambda df:[2, 3] ]
df.dropna()
```

```python

chi_tieu =  df['chi_tieu'].to_list
thu_nhap = df['thu_nhap'].to_list
```

```python
# adding the constant term
chi_tieu2 = sm.add_constant(chi_tieu)
 
# performing the regression
# and fitting the model
result = sm.OLS(thu_nhap, chi_tieu2).fit()
 
# printing the summary table
print(result.summary())
```
